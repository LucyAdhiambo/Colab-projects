```python
#13.Create a visualization to display the flipper_length_mm and body_mass_g for
### the pinguen dataset.
```

```python
plt.figure(figsize=(6,3))
plt.title('flipper_length_mm and body_mass_g')
sns.scatterplot(data=df,x='flipper_length_mm',y='body_mass_g',hue='species')
xlabel='flipper_length_mm'
ylabel='body_mass_g'
plt.show()
```

##### Obsevations
##### Scatter plot is another visualization tool,suitable when dealing with two or more numeric variables.
##### To interpret the relationship-each dot represents a penguin. The 'Adelie' species has a lower body_mass
##### and shorter flippers.'Gentoo' species are heavier and in-turn have longer flipper and 'Chinstrap' are in-between.
##### thus their is a linear relation between flipper_length and body_mass i.e there is a correlation.

```python
#14. Create a histogram with the flipper_length_mm and bins = 10
```

##### more easily and clearer.
